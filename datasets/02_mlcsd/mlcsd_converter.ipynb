```python
import pandas as pd
data_class_dataset_file_path = "./input/data_class_sld.csv"
god_class_dataset_file_path = "./input/god_class_sld.csv"

#! Open & Sort DataClass_Dataset
originalDataClassDf = pd.read_csv(
    data_class_dataset_file_path, on_bad_lines="skip").sort_values(by=[])

display(originalDataClassDf)

# print("originalDataClassDf Columns:", originalDataClassDf.columns)
print("originalDataClassDf Shape:", originalDataClassDf.shape)

#! Open & Sort GodClass_Dataset
originalGodClassDf = pd.read_csv(
    god_class_dataset_file_path, on_bad_lines="skip").sort_values(by=[])

display(originalGodClassDf)

# print("originalGodClassDf Columns:", originalGodClassDf.columns)
print("originalGodClassDf Shape:", originalGodClassDf.shape)


# Merge 2 CSV Datasets on the 'IDType' column (outer join)
merged_df = pd.merge(originalDataClassDf,
                     originalGodClassDf, on="IDType", how="outer")

# Remove Duplicated Columns
merged_df = merged_df.T.drop_duplicates().T

display(merged_df)
print(merged_df.shape)


data_class_instances_count = originalDataClassDf['is_data_class'].sum()
print("DataClass Instances Count:", data_class_instances_count)  # ! 140

god_class_instances_count = originalGodClassDf['is_god_class'].sum()
print("GodClass Instances Count:", god_class_instances_count)  # ! 140
```
